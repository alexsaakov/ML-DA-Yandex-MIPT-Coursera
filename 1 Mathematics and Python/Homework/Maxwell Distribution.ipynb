```python
%matplotlib inline
from scipy.stats import maxwell
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.stats import norm
import scipy.stats as sts
```

```python
# Создание случайной величины с распределением максвелла, зависящим
# от параметра alpha
alpha = 1.0
maxwell_rv = sts.maxwell(scale = alpha)

# Генерация выборки объёмом 1000 и создание гистограммы выборки
sample = maxwell_rv.rvs(1000)
plt.hist(sample, density=True, label='hist')

# Вычисление функции плотности распределения
x = np.linspace(0, 5, 100)
pdf = maxwell_rv.pdf(x)
plt.plot(x, pdf, label='pdf')

# Построение графика
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
x = np.linspace(0, 5, 100)
pdf = maxwell_rv.pdf(x)
plt.plot(x, pdf, 'k-', label='pdf')
```

```python
for n in [5, 10, 50, 100]:
    means = []
    for i in range(1000):
        sample = maxwell_rv.rvs(n)
        means.append(sample.mean())
    norm_rv = sts.norm(loc=2*alpha*np.sqrt(2/np.pi), scale=(((3*np.pi-8)/np.pi)*(alpha**2)/n)**0.5)    # Для распределения Максвелла: EX = 2*alpha*sqrt(2/pi), DX = (3*np.pi-8)*alpha^2/pi
    x = np.linspace(0, 5, 100)
    pdf = norm_rv.pdf(x)
    
    plt.hist(means, density=True, label='hist')
    plt.plot(x, pdf, label='norm pdf')
    
    plt.title('n = {}'.format(n))
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.legend()
    plt.show()
```

# Выводы

Точность аппроксимации распределения выборочных средних нормальным с ростом n улучшается, что можно наблюдать на графиках выше. Как видно, построенные графики являются наглядным подтверждением центральной предельной теоремы. Однако при слишком больших значениях n график сужается перестаёт быть информативным. Поэтому я считаю, что оптимальный размер n от 30 до 50.
