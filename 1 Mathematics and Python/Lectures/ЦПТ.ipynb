```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

## При выполнении задания использую теорему Эрдёша — Каца
### $\mu = log(log(n))$
### $\sigma = \sqrt{log(log(n))}$

```python
# определяю mu и sigma:
n = 1000
mu = np.log(np.log(n))
sigma = np.sqrt(mu)
```

```python
# задаю нормальное распределение, генерирую выборку из него
norm_rv = sts.norm(mu, sigma)
sample = norm_rv.rvs(1000)
```

```python
# строю гистограмму выборки и рисую теоретическую плотность распределения
plt.hist(sample, label='histogram', density=True)

x = np.linspace(min(sample)-3*sigma,max(sample)+3*sigma,200)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical PDF')
plt.legend()
plt.grid()
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

##  Pаспределение выборочного среднего при разных объёмах выборок

```python
# генерирую выборку размера n=5
a = []
mean_5 = []
for i in range(1000):
    a.append(norm_rv.rvs(5))
    mean_5.append(np.mean(a[i]))
```

```python
# строю гистограмму распределения выборочного среднего при n=5
sigma = np.sqrt(mu/5)
norm_rv5 = sts.norm(mu, sigma)
plt.hist(mean_5, bins=30, density=True, label='histogram')

# рисую плотность ee нормального распределения
x = np.linspace(min(mean_5)-3*sigma,max(mean_5)+3*sigma,100)
pdf_5 = norm_rv5.pdf(x)
plt.plot(x, pdf_5, label='theoretical PDF')
plt.legend()
plt.title('n=5')
plt.xlabel('$x$')
plt.ylabel('fraction of sample')
```

```python
# генерирую выборку размера n=10
b = []
mean_10 = []
for i in range(1000):
    b.append(norm_rv.rvs(10))
    mean_10.append(np.mean(b[i]))
```

```python
# строю гистограмму распределения выборочного среднего при n=10
sigma = np.sqrt(mu/10)
norm_rv10 = sts.norm(mu, sigma)
plt.hist(mean_10, bins=30, density=True, label='histogram')

# рисую плотность нормального распределения
x = np.linspace(min(mean_10)-3*sigma,max(mean_10)+3*sigma,100)
pdf_10 = norm_rv10.pdf(x)
plt.plot(x, pdf_10, label='theoretical PDF')
plt.legend()
plt.title('n=10')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
# строю гистограмму распределения выборочного среднего при n=50
sigma = np.sqrt(mu/50)
norm_rv50 = sts.norm(mu, sigma)
plt.hist(mean_50, bins=30, density=True, label='histogram')

# рисую плотность нормального распределения
x = np.linspace(min(mean_50)-3*sigma,max(mean_50)+3*sigma,100)
pdf_50 = norm_rv50.pdf(x)
plt.plot(x, pdf_50, label='theoretical PDF')
plt.legend()
plt.title('n=50')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
# генерирую выборку размера n=300
c = []
mean_300 = []
for i in range(1000):
    c.append(np.mean(norm_rv.rvs(300)))
    mean_300.append(np.mean(c[i]))
```
